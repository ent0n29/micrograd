```python
from micrograd.engine import Value
from trace_graph import draw_dot, trace 
from graphviz import Digraph
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# test data

# inputs x1,x2
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')
x3 = Value(3.0, label='x3')
x4 = Value(10.0, label='x4')
# weights w1,w2
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')
w3 = Value(12.0, label='w3')
w4 = Value(-4.0, label='w4')
# bias of the neuron
b = Value(6.8813735870195432, label='b')
# x1*w1 + x2*w2 + b
x1w1 = x1*w1; x1w1.label = 'x1w1'
x2w2 = x2*w2; x2w2.label = 'x2w2'
x3w3 = x3*w3; x3w3.label = 'x3w3'
x4w4 = x4*w4; x4w4.label = 'x4w4'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label = 'x1*w1 + x2*w2'
x3w3x4w4 = x3w3 + x4w4; x3w3x4w4.label = 'x3*w3 + x4*w4'
n = x1w1x2w2 + x3w3x4w4; n.label = 'n'
k = n + b; k.label = 'k' 
o = k.tanh(); o.label = 'o'
o.backward()
```
